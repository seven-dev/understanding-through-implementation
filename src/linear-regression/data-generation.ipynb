# Generate data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Generate time series data around a straight line

```python
np.random.seed(0)  # Setting seed for reproducibility
num_points = 100
x = np.linspace(0, 10, num_points)
m, c = 2, 5  # Slope and intercept for our straight line
y = m * x + c
noise = np.random.normal(0, 2, num_points)  # Random noise
y_with_noise = y + noise
```

## Display the line

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y_with_noise, color='blue', label='Data with Noise')
plt.plot(x, y, color='red', label='Original Line', linewidth=2)
plt.title('Time Series Data Around a Straight Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

## Load data into pandas & save as CSV

```python
df = pd.DataFrame({'X': x, 'Y': y_with_noise})
csv_path = './data.csv'
df.to_csv(csv_path, index=False)
```

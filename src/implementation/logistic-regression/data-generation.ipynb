# Generate data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Generate 2 variable data around a circle

```python
np.random.seed(0)  # Setting seed for reproducibility
num_points = 100
x_dist = 30
y_dist = 20

# Polar coordinates
theta = np.random.uniform(0,360,(num_points,))
R = np.random.uniform(0,3,(num_points,))
x_a = 2*R * np.cos(theta) + x_dist
y_a = R * np.sin(theta) + y_dist

# Polar coordinates
theta = np.random.uniform(0,360,(num_points,))
R = np.random.uniform(3,6,(num_points,))
x_b = 2*R * np.cos(theta) + x_dist
y_b = R * np.sin(theta) + y_dist
```

## Display the line

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_a, y_a, color='blue', label='Data of class A')
plt.scatter(x_b, y_b, color='red', label='Data of class B')
plt.title('Data around a circle')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.set_ylim(bottom=0)
ax.set_xlim(left=0)
plt.show()
```

## Load data into pandas & save as CSV

```python
# Stack a column of ones on x_a, y_a
a_class = np.hstack((x_a.reshape(x_a.shape[0], 1), y_a.reshape(y_a.shape[0], 1), np.ones((x_a.shape[0],1))))
# Stack a column of zeros on x_b, y_b
b_class = np.hstack((x_b.reshape(x_b.shape[0], 1), y_b.reshape(y_b.shape[0], 1), np.zeros((x_b.shape[0],1))))
# Store A and B classes in the same dataframe
all_classes = np.vstack((a_class, b_class))
df = pd.DataFrame(all_classes, columns = ['X','Y','Class'])
```
